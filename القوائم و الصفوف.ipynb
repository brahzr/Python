**1 . List**

```python
L = [12,15,18,12.5,"Exellent", True, 12, 12]
```

```python
type(L)
```

**1.2 List Slicing** استخراج القيم

```python
#extract 18 from the list L
L[2]
```

```python
#extract "exellent" from L
L[4]
```

```python
#extract 12 and True
print(L[0],L[5])
```

```python
#from 12.5 to True
L[3:6]
```

```python
L[3:]
```

**1.2 Changing items**

```python
L
```

```python
L[3] = 14.5
L[5] = False
L[4] = "Good"
print(L)
```

**1.3 List Functions**

```python
L.append(25)
print(L)
```

```python
L_copy=L.copy()
print(L_copy)
```

```python
L_copy.clear()
print(L_copy)
```

```python
L.count(12)
```

```python
M = [19,12.5]
L.extend(M)
print(L)
```

```python
L.remove(12)
print(L)
```

```python
L.pop(6)
print(L)
```
