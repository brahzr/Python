### If conditionals

*==* : مساواة  <br>
*<* : أقل <br>
*>* : أكبر <br>
*<=* : أقل أو يساوي <br>
*>=* : أكبر أو يساوي <br>
*!=* : لا تساوي

1. One condition (if)

```python
a = 8
b = 12
```

```python
if a<b:
    print("a is less than b")
```

```python
if a==b :
    print("a equals b")
```

```python
#example with list
c = [11,12,25]
d = [7.5,8.1,9.2]
if c[0] > d[0]:
    print("11 is bigger than 7.5")
```

2. Two and more conditions

```python
if a==b:
    print("a equals to b")
elif a>b:
    print("a is bigger than b")
elif a<b:
    print("a is less than b")
```

3. Else

```python
if a==b:
    print("a equals to b")
elif a>b:
    print("a is bigger than b")
else:
    print("a is less than b")
```

**Body Mass Index** <br>
Body mass index (BMI) is a measure of body fat based on height and weight in an individual, and then categorise that person as underweight, <br>
normal weight, overweight, or obese based on that value. <br>
The body mass index of a person with weight w (in kilograms) and height h (in metres) is calculated as: W/h**2 <br>
Use the above formula to compute the body mass index of someone who weighs 66.2 kilograms and has a height of 1.82 metres. <br>
**Print the result to the screen**.

```python
W = 95.2
h = 1.62
BMI = (W / h**2)
#obese BMI>=30
#overweight 25<BMI<30
#normalweight 19=<BMI<=25
#underweight <19
if BMI<19:
    print("This person is underweight")
elif 19<=BMI<=25:
    print("This person is normal weight")
elif 25<BMI<30:
    print("This person is overweight")
else:
    print("This person is Obese")
```

### AND, OR, NOT

**AND** <br>
TRUE and TRUE : TRUE <br>
TRUE and FALSE: FALSE <br>

```python
module1 = 12
module2 = 8
if module1>10 and module2>10:
    print("PASSED")
else:
    print("FAILED")
```

**OR** <br>
TRUE and TRUE : TRUE <br>
TRUE and FALSE: TRUE <br>

```python
module1 = 12
module2 = 8
if module1>10 or module2>10:
    print("PASSED")
else:
    print("FAILED")
```

**NOT** <br>
NOT TRUE : FALSE <br>
NOT FALSE : TRUE <br>

```python
module1 = 12
module2 = 8
if module1>10 and not module2>10:
    print("PASSED")
else:
    print("FAILED")
```

### For

```python
country = ["ALG", "TUN", "EGY"]
```

```python
for x in country:
    print(x)
```

**range function**

```python
for y in range(6):
    print(y)
```

1. The break Statement

```python
for x in country:
    print(x)
    if x == "TUN":
        break
```

2. The continue Statement

```python
for x in country:
    if x == "TUN":
        continue
    print(x)
```

3. Else in For Loop

```python
for x in country:
    print(x)
else:
    print("Items finished")
```

4. Pass Statement

```python
for x in [0, 1, 2]:
  pass
```

### While

The while loop requires relevant variables to be ready

```python
i = 2
while i < 6:
  print(i) #never stop
```

```python
i = 2
while i < 6:
  print(i)
  i += 1
```

1. The break Statement

```python
i = 1
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1
```

2. The continue Statement

```python
i = 1
while i < 6:
  i += 1
  if i == 5:
    continue
  print(i)
```

3. The else statement

```python
i = 1
while i < 6:
  print(i)

  i += 1
else:
  print("i is no longer less than 6")
```
