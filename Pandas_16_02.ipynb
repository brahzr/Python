```python
import matplotlib.pyplot as plt
```

```python
meat["Meat"].plot(title = "Meat Plot Line", xlabel = "Year", ylabel = "Meat Values",
                  color = "green", kind = "line")
plt.savefig("Meat Plot Line.png")
plt.show()
meat["Meat"].plot(title = "Meat Plot Boxplot", xlabel = "Year", ylabel = "Meat Values",
                  color = "green", kind = "box")

plt.savefig("Meat Plot Boxplot.png")
plt.show()
meat["Meat"].plot(title = "Meat Plot Histogramme", xlabel = "Year", ylabel = "Meat Values",
                  color = "green", kind = "hist")
plt.savefig("Meat Plot Hist.png")
plt.show()
meat["Meat"].plot(title = "Meat Plot Distribution", xlabel = "Year", ylabel = "Meat Values",
                  color = "green", kind = "kde")
plt.savefig("Meat Plot Dist.png")
plt.show()
```
