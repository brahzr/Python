### If conditionals

1. One condition (if)

```python
a = 8
b = 12
```

```python
if a==b :
    print("a equals b")
```

```python
if a<b:
    print("a is less than b")
```

```python
#list
c = [11,12,25]
d = [7.5,8.1,9.2]
if c[0] > d[0]:
    print("11 is bigger than 7.5")
```

2. Two and more conditions

```python
if a==b:
    print("a equals to b")
elif a>b:
    print("a is bigger than b")
elif a<b:
    print("a is less than b")
```

3. Else

```python
W = 95.2
h = 1.62
BMI = (W / h**2)
#obese BMI>=30
#overweight 25<BMI<30
#normalweight 19=<BMI<=25
#underweight <19
if BMI<19:
    print("This person is underweight")
elif 19<=BMI<=25:
    print("This person is normal weight")
elif 25<BMI<30:
    print("This person is overweight")
else:
    print("This person is Obese")
```

### AND, OR, NOT

```python
#AND
height1 = 1.62
height2 = 1.78
h_base = 1.75
if height1>h_base and height2>h_base:
    print("They are accepted")
else:
    print("They are not accepted")
```

```python
#OR
height1 = 1.62
height2 = 1.78
h_base = 1.75
if  height1>h_base or height2>h_base:
    print("They  are accepted")
else:
    print("They are not accepted")
```

```python
#Not
height1 = 1.62
height2 = 1.78
h_base = 1.75
if not height1>h_base and height2>h_base:
          print("They are accepted")
else:
    print("They are not accepted")
```
