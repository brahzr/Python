## Dictionary

### Dictionary Characters

1. Does not allow duplicates in Keys
2. Changeable
3. Ordered_Indexed

```python
Samir = {"Height": 1.75,
     "Weight": 72}
type(Samir)
```

```python
Course = {"module" : "Python",
          "IDE" : "Jupyter",
          "Hours" : 10,
         "Success": True,
         "Key5": 10}
print(Course)
```

**1. Change & Adding new items**

```python
Course["module"]= "Python"
print(Course)
#Adding new items
Course["Mark"] = "Good"
print(Course)
```

**2. Acessing items**

```python
D = {"Name": ["Ahmed", "Brahim", "Amine"],
     "Weight": [65,85,72]} #key, value
type(D)
```
